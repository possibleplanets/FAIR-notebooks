# inverse-fair example

Note the correspondence between forward and inverse is not quite one to one. Concentration-driven FaIR *isn't really a thing* - it's provided because everybody seems to want it, but results shouldn't be relied upon.

```python
import numpy as np
from fair.inverse import inverse_fair_scm
from fair.forward import fair_scm
from fair.RCPs import rcp85
import matplotlib.pyplot as pl
```

```python
C, F, T, lambda_eff, ohc, heatflux, af = fair_scm(
    emissions = rcp85.Emissions.emissions,
    gir_carbon_cycle=True,
    temperature_function = 'Geoffroy'
)
```

```python
pl.plot(np.arange(1765, 2501), T)
```

```python
pl.plot(np.arange(1765,2501), af)
```

```python
emissions, Fe, Te, lambda_eff, ohc, heatflux, afe = inverse_fair_scm(
    C = C[:,0],
    temperature_function = 'Geoffroy',
    F_in = np.sum(F, axis=1)
)
```

```python
pl.plot(np.arange(1765,2501), T)
pl.plot(np.arange(1765,2501), Te)
```

```python
pl.plot(np.arange(1765,2501), emissions)
```

```python
pl.plot(np.arange(1765, 2501), rcp85.Emissions.co2)
```

```python
pl.plot(np.arange(1765, 2501), afe)
pl.plot(np.arange(1765, 2501), af)
```
